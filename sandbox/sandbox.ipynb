```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.sin(x) + x

def f_dot(x):
    return np.cos(x) + 1

x = np.linspace(0, 20, 100)
y = f(x)
y_dot = f_dot(x)

plt.plot(x, y, label=" y = f(x) = sin(x) + x")
plt.plot(x, y_dot, label="y = f'(x) = cos(x) + 1")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.savefig("./drawings/stationarity.pdf", bbox_inches='tight')
```

```python
x = np.linspace(2, 15, 100)

y_1 = 2/np.log(x)
y_2 = 2/np.log(x) + x/10

plt.plot(y_1*10, label="Training Loss")
plt.plot(y_2*10, label="Validation Loss")
plt.vlines(29, ymin=5, ymax=28, colors='green', linestyles='dashed', label="Optimal Stopping Point")
plt.legend()
ax = plt.gca()
ax.set_ylim([5, 28])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.xticks([])
plt.yticks([])
plt.savefig("./drawings/early_stopping.pdf", bbox_inches='tight')
```
